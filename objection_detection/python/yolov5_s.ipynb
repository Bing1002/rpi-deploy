```python
import torch

device = 'cuda' if torch.cuda.is_available() else 'cpu'

# Load the YOLOv5s model from PyTorch Hub
model = torch.hub.load('ultralytics/yolov5:v6.2', 'yolov5s', 
                       pretrained=True, device=device, verbose=False)

# Set the model to evaluation mode
model.eval()
```

```python
model
```

```python
# Example image URL or path
img = '../../data/bus.jpg'  # Replace with your image path or URL

# Perform inference
results = model(img)

# Print results (bounding boxes, confidence scores, etc.)
print(results.pandas().xyxy)  # Print results as a pandas DataFrame
```
